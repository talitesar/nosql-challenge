## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
uk_food = mongo.uk_food
```

```python
# review the collections in our database
print(uk_food.list_collection_names())
```

```python
# assign the collection to a variable
establishments = uk_food.establishments
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':20}

# Use count_documents to display the number of documents in the result
print(f'There are {establishments.count_documents(query)} establishments with a hygiene score of 20.')

# Display the first document in the results using pprint
pprint(establishments.find(query)[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20 = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f'There are {len(hygiene_20)} rows in the DataFrame.')

# Display the first 10 rows of the DataFrame
hygiene_20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue': {'$gte':'4'}, 'LocalAuthorityName': {'$regex': 'London'}}

# Use count_documents to display the number of documents in the result
print(f'There are {establishments.count_documents(query)} establishments in London with a rating value greater than or equal to 4.')

# Display the first document in the results using pprint
pprint(establishments.find(query)[0])
```

```python
# Convert the result to a Pandas DataFrame
rating_4 = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f'There are {len(rating_4)} rows in the DataFrame.')

# Display the first 10 rows of the DataFrame
rating_4.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5

# Save variables for the search radius and Penang Flavours' latitude and longitude
degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

# Create a query where the rating value must be 5, hygiene score must not be empty, 
# and the latitude and longitude must be within 0.01 degree of Penang Flavours' coordiates using the stored values

query = {'RatingValue':'5',
        'geocode.latitude': {'$lte':latitude + degree_search},
        'geocode.latitude': {'$gte':latitude - degree_search},
        'geocode.longitude': {'$lte':longitude + degree_search},
        'geocode.longitude': {'$gte':longitude - degree_search},
        'scores.Hygiene': {'$ne': None}}

# Sort by hygiene score in ascending order so that lowest score (greatest hygiene) are first
sort =  [('scores.Hygiene',1)]

# Set a limit variable equal to 5 to grab the top five establishments
limit = 5

# Use these conditions in a search, convert it into a list, and save it in a variable
search = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
for result in search:  
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
search_df = pd.DataFrame(search)

# Display DataFrame
search_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match':{'scores.Hygiene':0}}
# 2. Groups the matches by Local Authority
group_query = {'$group':
                {'_id':'$LocalAuthorityName',
                'count': {'$sum':1}}}
# 3. Sorts the matches from highest to lowest
sort_query = {'$sort': {'count':-1}}

# Put the elements of the pipeline together
pipeline = [match_query, group_query, sort_query]

# Save the reuslts of this pipeline to a list
hygiene_count = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f'There are {len(hygiene_count)} documents in this result.')

# Print the first 10 results
pprint(hygiene_count[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hygiene_count)
# Display the number of rows in the DataFrame
print(f"There are {len(hygiene_df)} rows in the DataFrame.")
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```
